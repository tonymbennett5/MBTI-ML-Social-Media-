## Count Vectorizer 

A count vectorizer encodes text by splitting a set of words into one column per word , with the count of the word for that row into the column 

```python
# vectorizer with defaul settings 
vec = CountVectorizer()
vec.fit_transform(df_clean2['cleaned posts'])
```

```python
# output of the transformation to a dense matrix 
vec.transform(df_clean2['cleaned posts']).todense()
```

```python
# Convert to a dataframe to get feature names and see the distribution of words
cleaned_posts = pd.DataFrame(vec.transform(df_clean2['cleaned posts']).todense(), columns=vec.get_feature_names())
```

```python
cleaned_posts.head()
```

```python
cleaned_posts.shape
```

- created 96376 columns from the cleaned posts column

```python
# top the top 10 occuring words
print(cleaned_posts.sum().sort_values(ascending=False).head(10))
```

```python
# rerunning vectorizer now allowing two word pairs and having only 1000 columnns 
vec = CountVectorizer(ngram_range = (1,2), max_features = 1000)
vec.fit(df_clean2['cleaned posts'])
vec.transform(df_clean2['cleaned posts'])

vec.transform(df_clean2['cleaned posts']).todense()
cleaned_posts = pd.DataFrame(vec.transform(df_clean2['cleaned posts']).todense(), columns=vec.get_feature_names())
```
